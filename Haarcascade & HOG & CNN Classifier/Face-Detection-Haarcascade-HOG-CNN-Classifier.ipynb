## Detecting Faces

```python
face_detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
detections = face_detector.detectMultiScale(image_gray)
```

```python
detections
```

```python
for (x,y,w,h) in detections:
    #print(x,y,w,h)
    cv2.rectangle(image, (x,y), (x+w,y+h), (255,255,0), 2)

plt.imshow(image)
#cv2.imshow('Detected',image)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

## Improve False Postive bounding box with Haarcascade parameters

```python
image = cv2.imread('images/faces_001.jpg')
#image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.resize(image, (800,600))
image_gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
face_detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
detections = face_detector.detectMultiScale(image_gray, scaleFactor=1.45)  # parameter passed
for (x,y,w,h) in detections:
    cv2.rectangle(image, (x,y), (x+w, y+h), (0,255,0), 2)
plt.imshow(image)

cv2.imshow('Detected',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.imread('images/faces_002.jpg')
#image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.resize(image, (800,600))
image_gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
face_detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
#detections = face_detector.detectMultiScale(image_gray, scaleFactor=1.45, minNeighbors=6, minSize=(14,14), maxSize=(100,100)) # parameter passed
face_detections = face_detector.detectMultiScale(image_gray, scaleFactor=1.1) # parameter passed
for (x,y,w,h) in face_detections:
    #print(w,h)
    cv2.rectangle(image, (x,y), (x+w, y+h), (0,255,0), 2)
plt.imshow(image)

cv2.imshow('Detected',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## The dlib library provides two functions that can be used for face detection:

## Face Detection using HOG

```python
import dlib
```

```python
image = cv2.imread('images/faces_005.jpg')
#image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.resize(image, (800,600))

cv2.imshow('Image',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
face_detector_hog = dlib.get_frontal_face_detector()
```

```python
detections = face_detector_hog(image,1)
```

```python
detections
```

```python
# Number of faces detected
len(detections)
```

```python
for face in detections:
    #print(face)
    l, t, r, b = face.left(), face.top(), face.right(), face.bottom()
    cv2.rectangle(image, (l,t), (r,b), (255,0,0),2)
    

plt.imshow(image)    
cv2.imshow('Image',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Face Detection using CNN

```python
import dlib
```

```python
image = cv2.imread('images/faces_005.jpg')
#image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.resize(image, (800,600))

cv2.imshow('Image',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# load pre-trained Max-Margin (MMOD) CNN face detector model
cnn_detector = dlib.cnn_face_detection_model_v1('mmod_human_face_detector.dat')
```

```python
detections = cnn_detector(image,1)
```

```python
for face in detections:
    #print(face)
    l, t, r, b, c = face.rect.left(), face.rect.top(), face.rect.right(), face.rect.bottom(), face.confidence
    print(c) # Confidence score
    cv2.rectangle(image, (l,t), (r,b), (255,0,0),2)
    

plt.imshow(image)    
cv2.imshow('Image',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
